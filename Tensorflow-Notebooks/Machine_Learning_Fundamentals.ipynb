Artificial Intelligence (A.I.):
*The effort to automate intellectual tasks normally performed by humans*
***
* 1950: Question asked: Can we get a computer to think and do its own thing?
* Back then A.I. was simply a pre-defined set of rules generated by humans and implemented in code that computers would then execute and follow 
* No crazy machine learning algorithms or deep learning yet 
* Very good A.I. was simply just a very good set of rules
* **A.I. can be simple or complex**


Machine Learning (ML):
*Rather than giving the program the rules, an algorithm finds the rules for us*
***
* **Machine Learning is a part of Artificial Intelligence**
* Machine Learning figures out the rules for us
* Takes the input data, takes the expected output and figures out what rules to generate so that when we look at new data, we can have the best possible output for that
* Attempts to come up with rules for a dataset
* This is accomplished by feeding in some features and labels to a model (training data) and tell it to start training and this model will then come up with rules such that we can just give some features to the model in the future and it should be able to give us a pretty good estimate of what the label should be
* Different types of techniques including unsupervised, supervised, & reinforcement learning
* **Goal for machine learning model: highest accuracy possible**

Neural Networks/Deep Learning:
*A form of machine learning that uses a layered representation of data*
***
* We have an input layer (first layer of data) 
* We could have some layers in between this layer that are all connected together
* Finally have some output layer
* **Essentially our data will be transformed through different layers, and evetually we will reach some output**
* Sometimes referred to as a multi stage information extraction process
* Multiple layers of information whereas in standard achine learning, we only have 1 or 2 layers

The importance of data: *The key thing we use to create models whether we are doing A.I. or Machine Learning*
***
* Whatever input information we have that we give to the model will be our ***features***
* ***Features*** are what we need to give to our model to train it to make a prediction
* The output of the model is what is known as ***labels***
* ***Labels*** are simply what we are trying to look for or predict
***When we feed our features to a model, it will give to us a label***

Different types of Machine Learning: 
***
  * Unsupervised Learning
  * Supervised Learning
  * Reinforcement Learning




Supervised Learning:
***
* We have some features and corresponding label(s)
* We pass the into some machine learning model that figures out the rules for us
* Later on, all we do is give it some features and the model will then give us the labels using the rules it generated
* **Reason why it's called supervised: When we pass in the features to oure model, it makes some arbitrary prediction using the rules it already knows and then it compares the label(s) it predicted to what the actual label(s) are --> so it supervises the model and tweaks the model for improvement through several 100s or even 100s of times until it eventually becomes solid enough that we can stop training**
* Most common and most applicable type of learning 
* **Massive advantage when you have a ton of features and corresponding label(s)**

Unsupervised Learning:
***
* When we only have features available (absolutely no labels) 
* What we want to do is have our model come up with those labels for us
* ***Don't have our label(s), we want our model to find the label(s) for us, it doesn't matter how we get there as long as we get there***
* A good example of this may be when clustering data point and trying to find the predicted labels for new data points 

Reinforcement Learning:
***
* We don't have any data, we instead have what is called **an agent, an environment, and a reward**
* The objective of the agent is to maximize its reward and a reward is essentially what the agent gets when it does something correctly
* Start off with an agent who knows absolutely nothing about the environment, and it starts exploring (both randomly and by using knowledge it has gathered before) to try to maximize its reward
* This makes it so that in the future, when we plug in the agent into the environment again, it will know how to maximize its reward whilst completing its objective
* We, the programmers, need to find a way to reward the agent correctly so that it gets to the objective in the best possible way
* Real application of this is training AI's to play games
